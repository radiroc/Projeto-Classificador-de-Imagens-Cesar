```python
from bing_image_downloader import downloader

def get_images(query):
    """Download images and place them in a directory.
    
    :param query: Query to search for
    :return: Images matching query
    
    """
    print(query)
    
    downloader.download(query, 
                    limit=360, 
                    output_dir='home', 
                    adult_filter_off=False, 
                    force_replace=False, 
                    timeout=60)
```

```python
humans = [
'Adult Person GettyImages',
'Child',
'Elderly', #idoso
]

```

```python
for human in humans:
    print('Fetching images of', human)
    get_images(human)
```

```python
import os

os.rename('home/Adult Person GettyImages', 'home/Adult')


origem = '/home'
humans = 'Humans'

path = os.path.join(origem, humans)

os.mkdir(path)

```

```python
end = '/home/Humans'

adult = 'Adult'
kid = 'Child'
elderly = 'Elderly'

def cria_dir(classe):
  path_f = os.path.join(end,classe)

  os.makedirs(path_f)

  train = 'train'
  valid = 'val'

  end_f = end + '/' + classe

  path_train = os.path.join(end_f, train)
  path_val = os.path.join(end_f,valid)

  os.makedirs(path_train)
  os.makedirs(path_val)


cria_dir(adult)
cria_dir(kid)
cria_dir(elderly)

```

```python

import os
import shutil

data_a = os.listdir('/content/home/Adult')
data_k = os.listdir('/content/home/Child')
data_e = os.listdir('/content/home/Elderly')

def insert_files (data,classe):
  from sklearn.model_selection import train_test_split
  train, valid = train_test_split(data, test_size=0.3, random_state=1)
  
  dest_tr = '/home/Humans/'+ classe + '/train'
  dest_val = '/home/Humans/'+ classe + '/val'

  for i in train:
    orig_tr = '/content/home/'+ classe + '/' + i
    shutil.move(orig_tr,dest_tr)
  for v in valid: 
    orig_val = '/content/home/'+ classe + '/' + v
    shutil.move(orig_val,dest_val)
```

```python
insert_files(data_a,adult)
insert_files(data_k,kid)
insert_files(data_e,elderly)
```
